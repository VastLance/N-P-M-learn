# Seaborn 学习笔记

```python
# 导库
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 导入数据
penguins = pd.read_csv('data/penguins.csv')

# penguins.head()
# penguins.info()

# 缺失值处理
penguins.dropna(inplace=True)

penguins.head()
```

## 直方图

```python
# 直方图
sns.histplot(data=penguins,
             x='species',
             hue='sex',
             # multiple='stack'
             )
```

## 核密度估计图 KDE

```python
# 核密度估计图 （KDE）
sns.kdeplot(data=penguins,
            x='body_mass_g',
            hue='species',
            # multiple='stack'
            )
```

## 直方图+KDE

```python
# 核密度估计图 （KDE）
sns.histplot(data=penguins, x='body_mass_g', hue='species', kde=True)
```

## 计数图

```python
# 计数图
sns.countplot(data=penguins, x='species', hue='sex')
```

## 散点图

```python
# 散点图
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

## 散点图+KDE

```python
# 散点图 + KDE
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

## 蜂窝图

```python
# 蜂窝图
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='hex')
```

## 蜂窝图+KDE

```python
# 蜂窝图 + KDE
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', kind='kde')
```

## 回归图

```python
# 回归图
sns.lmplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

## 二维核密度估计图（2-D KDE）

```python
# 二维核密度估计图（2-D KDE）
sns.kdeplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', fill=True, cbar=True)
```

## 条形图

```python
# 条形图
sns.barplot(data=penguins, x='species', y='bill_length_mm', estimator='mean', errorbar=None)
```

## 箱线图

```python
# 箱线图
sns.boxplot(data=penguins, x='species', y='bill_length_mm')
```

## 小提琴图

```python
# 小提琴图
sns.violinplot(data=penguins, x='species', y='bill_length_mm')
```

## 成对关系图

```python
# 成对关系图
sns.pairplot(data=penguins, hue='species')
```
