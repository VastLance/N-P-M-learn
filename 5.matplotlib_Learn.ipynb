```python
#导入库
import matplotlib.pyplot as plt
from matplotlib import rcParams
# 设置字体
rcParams['font.family'] = 'SimHei' # 黑体
rcParams['axes.unicode_minus'] = False # 正常显示负号
```

## 折线图

```python
# 创建图表，设置大小
plt.figure(figsize = (5, 3)) # 6 是宽度，3 是高度

# 要绘图的数据
month = ['1月','2月','3月','4月','5月']
sales = [100, 130, 120,150, 160]

# 绘制折线图
plt.plot(month, sales,
         marker = 'o',
         color = 'g',
         linestyle = '-',
         linewidth = 3,
         markersize = 5,
         label = '产品A')

# 添加标题
plt.title('2025年前5月各月份销售情况', color='red', fontsize=15)

# 添加x轴和y轴标签
plt.xlabel('月份', color='y', fontsize=10)
plt.ylabel('销售额 / 万元', color='y', fontsize=10)

# 设置刻度字体大小
plt.xticks(fontsize=10, rotation=0)
plt.yticks(fontsize=10, rotation=0)

# 设置x轴和y轴的范围
plt.xlim(0, 4)
plt.ylim(0, 180)

# 添加图例
plt.legend(loc = 'upper left') # 图例位置: 左上角

# 显示网格
plt.grid(True, linestyle = '--', color = 'gray', alpha = 0.5) # 显示网格，网格线样式为虚线，颜色为灰色，透明度为0.5
# plt.grid(axis='y',alpha=1) # 只显示y轴的网格线，透明度为1

# 在每个数据点上显示数值
for x,y in zip(month, sales):
    plt.text(x, y+6, y, ha='center', va='bottom', fontsize=10)

# 自动排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 绘制多个折线图
# 创建图表，设置大小
plt.figure(figsize = (5, 3)) # 5 是宽度，3 是高度
# 要绘图的数据
month = ['1月','2月','3月','4月','5月']
sales1 = [100, 130, 120,150, 160]
sales2 = [120, 140, 130,160, 170]
sales3 = [110, 135, 125,145, 155]
# 绘制折线图
plt.plot(month, sales1,
         marker = 'o',
         color = 'g',
         linestyle = '-',
         linewidth = 3,
         markersize = 5,
         label = '产品A')
plt.plot(month, sales2,
         marker = 'o',
         color = 'r',
         linestyle = '-',
         linewidth = 3,
         markersize = 5,
         label = '产品B')
plt.plot(month, sales3,
         marker = 'o',
         color = 'b',
         linestyle = '-',
         linewidth = 3,
         markersize = 5,
         label = '产品C')
plt.legend(loc = 'upper left') # 图例位置: 左上角
plt.show()
```

## 柱状图

```python
# 创建图表，设置大小
plt.figure(figsize = (5, 3)) # 5 是宽度，3 是高度

# 要绘图的数据
subjects = ['语文','数学','英语','物理','化学']
scores = [130, 150, 145, 100, 100]
# 绘制柱状图
plt.bar(
    x = subjects,
    height = scores,
    color = 'orange',
    alpha = 1,
    label = 'Ada',
    width = 0.5
)

plt.legend(loc='upper left')

plt.xlabel('科目')
plt.ylabel('分数')
plt.title('成绩')

plt.grid(axis='y',alpha=0.3, color='gray')

for x,y in zip(subjects, scores):
    plt.text(x,y+5,y,ha='center',va='bottom',fontsize=10)

plt.tight_layout()  # 自动排版

plt.show()
```

## 条形图

```python
# 创建图表，设置大小
plt.figure(figsize = (7, 3)) # 5 是宽度，3 是高度
# 要绘图的数据
GDP = [140, 150, 110, 110, 130]
Country = ['中国','美国','日本','韩国','英国']
# 绘制条形图
plt.barh(
    y = Country,
    width = GDP,
    color = 'orange',
    alpha = 1,
    label = 'GDP'
)
plt.legend(loc='upper left')
plt.xlabel('GDP')
plt.ylabel('国家')
plt.xlim(0,170)
plt.title('国家GDP')
plt.grid(axis='x',alpha=0.3, color='gray')
for x,y in zip(GDP, Country):
    plt.text(x,y,x,ha='left',va='center',fontsize=10)

plt.tight_layout()
plt.show()
```

## 饼图

```python
plt.figure(figsize = (5, 4))
things = ['study', 'relex', 'eat', 'sleep', 'exercise']
time = [10, 2, 3, 8, 1]
plt.pie(
    time,  # 分隔依据
    labels = things,  # 标签
    autopct = '%.2f%%',  # 显示百分比，保留两位小数
    shadow = True,  # 显示阴影
    startangle = 90,  # 起始角度
    colors = ['#63d5a1','#65cccc','blue','yellow','orange'],  # 颜色，可以是颜色名称，也可以是颜色16进制代码
    explode = [0.15,0,0,0,0]  # 突出显示（爆炸效果）
)
plt.title('生活时间')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

## 环形图

```python
plt.figure(figsize = (5, 4))
things = ['study','relex', 'eat','sleep', 'exercise']
time = [10, 2, 3, 8, 1]
plt.pie(
    time,
    labels = things,
    radius = 1, # 环形图的半径
    counterclock = False, # True:逆时针方向, False:顺时针方向
    autopct = '%.2f%%',
    shadow = True,
    startangle = 90,
    colors = ['#84d3a4','#62c255','blue','yellow','orange'],
    explode = [0.1,0,0,0,0],
    wedgeprops = {'width':0.6}, # 环形图的宽度
    pctdistance = 0.7 # 百分比标签的距离
)
plt.text(0,0,'total',ha='center',va='center',fontsize=15)
plt.title('time distribution')
plt.show()
```

## 散点图

```python
import random
from numpy import arange

plt.figure(figsize = (4, 3))
random.seed(10)
x = [random.randint(0,20) for i in range(30)]
y = [(x[i] + random.randint(-3,3)) for i in range(30)]

plt.scatter(
    x=x,y=y,
    marker='*',
    color='#03b953',
    alpha=0.6,
    label='data',
    s=40 # 点的大小
)

plt.plot(
    arange(0,20,1),
    arange(0,20,1),
    color='red',
    linestyle='--',
    alpha=0.6,
    linewidth=2,
    label='y = x'
)

plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

## 箱线图

```python
plt.figure(figsize = (4, 3))

data = {
    'science': [100, 130, 120, 150, 160, 150, 180, 130],
   'art': [120, 140, 130, 160, 170, 60, 100, 130],
   'english': [110, 135, 125, 145, 155, 180, 135, 145]
}
plt.boxplot(
    data.values(),
    tick_labels=data.keys(),
    # showmeans=True,
    # meanline=True,
    # notch=True,
    # patch_artist=True
)
plt.title('成绩分布')
plt.xlabel('subjects',color='red',fontsize=15)
plt.ylabel('scores',color='red',fontsize=15)
plt.grid(axis='y',alpha=0.3,color='gray',linestyle='--')
plt.tight_layout()
plt.ylim(30,210)
plt.xticks(fontsize=12,rotation=0)
plt.show()
```

## 多个图表的生成（子图）

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales1 = [100, 130, 120, 150, 160]
sales2 = [120, 140, 130, 160, 170]
sales3 = [110, 135, 125, 145, 155]
sales4 = [105, 120, 110, 130, 140]

f1 = plt.subplot(2, 2, 1)  # subplot是子图，（行，列，索引）
f1.plot(days,sales1)
f1.set_title('Product A')

# f2 = plt.subplot(2, 2, 2)
f2 = plt.subplot(222)  # 这样写也是可以的
f2.scatter(days,sales2)
f2.set_title('Product B')

f3 = plt.subplot(223)
f3.barh(days,sales3)
f3.set_title('Product C')

f4 = plt.subplot(2, 2, 4)
f4.pie(sales4,labels=days)
f4.set_title('Product D')

plt.tight_layout()
```

## 其他图表

### 直方图

```python
# 直方图
plt.figure(figsize = (5, 3))
data = [100, 130, 120, 150, 160, 150, 180, 130]
plt.hist(
    data,
    bins=5,
    color='#03b953',
    alpha=0.6,
    label='data',
)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

### 热力图

```python
# 热力图
import numpy as np
plt.figure(figsize = (5, 3))
data = np.random.rand(5, 5)
plt.imshow(
    data,
    cmap='hot',
    interpolation='nearest',
    aspect='auto'
)
plt.colorbar()
plt.tight_layout()
plt.show()
```

### 面积图

```python
# 面积图
plt.figure(figsize = (5, 3))
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales1 = [80, 120, 110, 130, 150]
sales2 = [120, 140, 130, 160, 170]
plt.fill_between(
    days,
    sales1,
    color='#03b953',
    alpha=0.6,
    label='Product A'
)
plt.fill_between(
    days,
    sales2,
    color='#65cccc',
    alpha=0.7,
    label='Product B'
)
plt.legend(loc='upper left')
plt.show()
```

### 堆叠图

```python
# 堆叠图
plt.figure(figsize = (5, 3))
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales1 = [80, 120, 110, 130, 150]
sales2 = [120, 140, 130, 160, 170]
sales3 = [110, 135, 125, 145, 155]
plt.stackplot(
    days,
    sales1,
    sales2,
    sales3,
    labels=['Product A', 'Product B', 'Product C']
)
plt.legend(loc='upper left')
plt.show()
```

## 综合案例：温度分析

```python
# 1.导入库
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei' # 黑体
rcParams['axes.unicode_minus'] = False # 正常显示负号

# 2.读取数据
df = pd.read_csv('./data/weather.csv')
# df.head()

# 绘制气温的趋势变化图
df['date'] = pd.to_datetime(df['date'])
df = df[ df['date'].dt.year == 2015 ]
df['temp_mean'] = (df['temp_max'] + df['temp_min']) / 2
plt.figure(figsize = (10, 5))
plt.plot(df['date'], df['temp_max'],label='最高气温',alpha=0.6)
plt.plot(df['date'], df['temp_min'],label='最低气温',alpha=0.6)
plt.plot(df['date'], df['temp_mean'],label='平均气温',alpha=0.8)
plt.legend(loc='upper left')
plt.title('2015年气温趋势变化',fontsize=15)
plt.xlabel('日期',fontsize=12)
plt.ylabel('气温',fontsize=12)
```

```python
# 绘制降水量的直方图
plt.hist(df['precipitation'],bins=5,color='#0353b9',alpha=0.6,label='降水量')
plt.legend(loc='upper right')
plt.title('2015年降水量分布',fontsize=15)
plt.xlabel('降水量',fontsize=12)
plt.ylabel('频率',fontsize=12)
plt.show()
```
